```python
import tensorflow as tf
from tensorflow.keras import models , layers
import matplotlib.pyplot as plt
```

```python
def get_dataset_partitions(dataset,train_split=0.8,val_split=0.1,test_split=0.1,shuffle=True,shuffle_size=100000):
    # pass
    dataset_size = len(dataset)

    if shuffle :
        dataset = dataset.shuffle(shuffle_size,seed=12)

    train_size = int(train_split*dataset_size)
    val_size = int(val_split*dataset_size)


    train_ds = dataset.take(train_size)


    val_ds = dataset.skip(train_size).take(val_size)
    test_dataset = dataset.skip(train_size).skip(val_size)

    return train_ds,val_ds,test_ds
```

```python
train_ds,val_ds , test_ds = get_dataset_partitions(dataset)
```

```python
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds = val_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
test_ds = test_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
resize = tf.keras.Sequential([
    layers.experimental.preprocessing.Resizing(100,100),
    layers.experimental.preprocessing.Rescaling(1./255)
])
```

```python
data = tf.keras.Sequential([
    layers.experimental.preprocessing.RandomFlip('horizontal_and_vertical'),
    layers.experimental.preprocessing.RandomRotation(0.2)
])
```

```python
# train_ds = train_ds.map(
#     lambda x, y: (data(x, training=True), y)
# ).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
input_shape1 = (BATCH_SIZE, IMAGE_SIZE, IMAGE_SIZE, CHANNELS)
n_classes = 3
model = models.Sequential([
    resize,
    layers.Conv2D(32, kernel_size = (3,3), activation='relu', input_shape=input_shape1),
    layers.MaxPooling2D((2, 2)),
    # layers.Conv2D(64,  kernel_size = (3,3), activation='relu'),
    # layers.MaxPooling2D((2, 2)),
    # layers.Conv2D(64,  kernel_size = (3,3), activation='relu'),
    # layers.MaxPooling2D((2, 2)),
    # layers.Conv2D(64, (3, 3), activation='relu'),
    # layers.MaxPooling2D((2, 2)),
    # layers.Conv2D(64, (3, 3), activation='relu'),
    # layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(n_classes, activation='softmax'),
])

model.build(input_shape=input_shape1)
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=['accuracy']
)
```

```python
for batch in train_ds.take(1):  # Takes the first batch
    input_data, target = batch
    print(input_data.shape)
```

```python
history = model.fit(
    train_ds,  # Your training dataset
    batch_size=32,  # Adjust the batch size as needed
    validation_data=val_ds,  # Your validation dataset
    verbose=1,  # Set the verbosity level as needed (0, 1, or 2)
    epochs=50,  # Number of training epochs
)
```

```python
scores = model.evaluate(test_ds)
```
